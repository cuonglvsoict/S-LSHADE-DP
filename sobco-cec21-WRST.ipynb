```python
import sys
!{sys.executable} -m pip install --upgrade scipy

from scipy.stats import ranksums
import numpy as np

ALPHA = 0.05
codes = ["000", "010", "110", "011", "111"]

sub_path = [
    
            "cec21/S-LSHADE-DP/S-LSHADE-DP_",
#             "cec21/IMODE/IMODE_",
#             "cec21/j21/j21_",
            "cec21/NL-SHADE-RSP/NL-SHADE-RSP_",    
            ]

print(sub_path)
```

```python
overall = [[0,0,0] for i in range(len(sub_path)-1)]
for code in codes:
    print(code)
    count_sign = [[0,0,0] for i in range(len(sub_path)-1)]
    for i in range(1, 11):
        data = []
        for algo in sub_path:
            with open(algo + "({})_{}_10.txt".format(code, i)) as f:
                arr = [[float (x) for x in line.split()] for line in f]
                
                if algo == "cec21/S-LSHADE-DP/S-LSHADE-DP_":
                    data.append(arr[-2])
                    for idx in range(0, len(data[-1])):
                        if data[-1][idx] <= 1e-8:
                            data[-1][idx] = 0
                elif algo == "cec21/APGSK_IMODE/APGSK_IMODE_":
                    data.append(arr[-2])
                else:
                    data.append(arr[-1])


        pivot = np.mean(data[0])
        print("f{}\t{:.2E}\t{:.2E} ({:.2E})".format(i, np.min(data[0]), pivot, np.std(data[0])), end="")

        for j in range(1, len(data)):
            w, p = ranksums(data[0], data[j])
            avg = np.mean(data[j])

            sign = ""
            if p > ALPHA:
                sign = "="
                count_sign[j-1][0]+=1
                overall[j-1][0]+=1
            elif pivot < avg:
                sign = "+"
                count_sign[j-1][1]+=1
                overall[j-1][1]+=1
            else:
                sign = "-"
                count_sign[j-1][2]+=1
                overall[j-1][2]+=1


            print("\t{:.2E}\t{:.2E} ({:.2E}) {}".format(np.min(data[j]), avg, np.std(data[j]), sign), end="")

        print()
    print()

#     print("+/=/-\t\t", end="")
#     for j in range(1, len(sub_path)):
#         print("{}/{}/{}\t\t".format(count_sign[j-1][1], count_sign[j-1][0], count_sign[j-1][2]), end="")
#     print()
    
    
print("Overall performance")
print("+/=/-\t\t", end="")
for j in range(1, len(sub_path)):
    print("{}/{}/{}\t\t".format(overall[j-1][1], overall[j-1][0], overall[j-1][2]), end="")
print()
```
