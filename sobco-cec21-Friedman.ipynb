```python
from scipy.stats import friedmanchisquare
import numpy as np

ALPHA = 0.05

codes = ["000", "010", "110", "011", "111"]

sub_path = [
    
            "cec21/S-LSHADE-DP/S-LSHADE-DP_",
            "cec21/IMODE/IMODE_",
            "cec21/j21/j21_",
            "cec21/NL-SHADE-RSP/NL-SHADE-RSP_",    
            ]

print(sub_path)
```

```python
data = []
for algo in sub_path:
    mean = []
    for code in codes:
        for i in range(1, 11):
            with open(algo + "({})_{}_10.txt".format(code, i)) as f:
                arr = [[float (x) for x in line.split()] for line in f]
                
                temp = []
                if algo == "cec21/S-LSHADE-DP/S-LSHADE-DP_":
                    temp.extend(arr[-2])
                    
                    for idx in range(0, len(temp)):
                        if temp[idx] <= 1e-8:
                            temp[idx] = 0
                else:
                    temp.extend(arr[-1])
                
                mean.append(np.mean(temp))
    
    data.append(mean)
    
for arr in data:
    for item in arr:
        print(item, end="\n")
    print()
print(friedmanchisquare(data[0], data[1], data[2], data[3]))
```
